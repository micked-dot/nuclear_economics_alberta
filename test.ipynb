```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def simulate_gas_price(start=3.5, mu=3.5, sigma=0.5, years=40): #start=current price, mu=average price
    prices = [start]
    for _ in range(years - 1):
        shock = np.random.normal(0, sigma) #sigma of 0.5 indicates that 68% of annual shocks were within $0.5
        # will need to calibrate this. Get historical data, calculate the price changes, then do np.std(price_changes)
        # can also do scenarios/regime changes for that
        price = max(1.0, prices[-1] + 0.2 * (mu - prices[-1]) + shock) #not sure 0.2 is reasonable
        # linear regression to calculate speed of mu regression, but can also run models with a few different scenarios
        prices.append(price)
    return np.array(prices)

for n in range (1,10):
    gas_price = simulate_gas_price()
    plt.plot(gas_price)

plt.figure(figsize=(10, 8))
plt.plot(gas_price)
```
